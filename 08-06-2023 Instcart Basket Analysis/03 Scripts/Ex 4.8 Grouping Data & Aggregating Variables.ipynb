```python
# Step 4: create a loyalty flag for existing customers using transform() and loc() fun
# Split the data and apply transform()
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check the frequency
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Step 5: Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

#### The mean for Loyal customers is less than new and regular customers. Min and Max prices remain the same for all

```python
# Step 6: Create a spending flag for each user based on the average price across all their orders
# Split the data in groups based on user_id column and apply transform 
ords_prods_merge['mean_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Check the frequency of splending flags for accuracy
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Step 7: Create an order frequency flag that marks the regularity of a user’s ordering behavior
# Split and group based on user_id column and use transform to get the median for the purchase on days
ords_prods_merge['median_days'] = ords_prods_merge.groupby(['user_id'])['days_after_last_order'].transform(np.median)
```

```python
# Create flags to assign the frequency labels based on user ID on median days since prior order(days after last order)
ords_prods_merge.loc[ords_prods_merge['median_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days'] > 10) & (ords_prods_merge['median_days'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check the frequency for accuracy
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```
